```python
# dependencies
import pandas as pd
import numpy as np
from datetime import datetime

# file path setting

file_path = 'Resources/underemployed_data.xlsx'

# 
dateCols = ['Unnamed: 0']
# Read in all sheets
df = pd.read_excel(file_path, engine='openpyxl', sheet_name='Data2', parse_dates=dateCols)


# # # rename columns
df.rename(columns={'Unnamed: 0':'datetime_col'}, inplace=True)

# # drop unuseful rows
df = df.drop([0,1,2,3,4,5,6,7,8])

#remove NaNs
# df = df.dropna(axis=1, how='any', thresh=None, subset=None, inplace=False)

# # select time range
start_date = '2015-01-01 00:00:00'
end_date = '2020-12-01 00:00:00'
time_range_df = df.loc[(df['datetime_col'] >= start_date) & (df['datetime_col'] <= end_date)]

# time_range_df.columns

#select columns containing data relevant to Underemployment. 
unemployment_df = time_range_df.filter(like='Unemployed')


# selecting the columns with the "seasonally Adjusted" series type. 
unemployment_df1 = unemployment_df.filter(like=';.1', axis=1)
unemployment_df1
```

```python
# 
dateCols = ['Unnamed: 0']
# Read in all sheets
df2 = pd.read_excel(file_path, engine='openpyxl', sheet_name='Data3', parse_dates=dateCols)



# # rename columns
df2.rename(columns={'Unnamed: 0':'datetime_col'}, inplace=True)


# # drop unuseful rows
df2 = df2.drop([0,1,2,3,4,5,6,7,8])

# df2 = df2.drop(df2.columns[[4,5,6,7,8,9,10]] ,axis=1, inplace=True)


# # select time range
start_date = '2015-01-01 00:00:00'
end_date = '2020-12-01 00:00:00'
time_range_df2 = df2.loc[(df2['datetime_col'] >= start_date) & (df2['datetime_col'] <= end_date)]

# # time_range_df.columns

# #select columns containing data relevant to Underemployment. 
unemployment_df2 = time_range_df2.filter(like='Unemployed')


# # selecting the columns with the "seasonally Adjusted" series type. 
unemployment_df2 = unemployment_df2.filter(like=';.1', axis=1)
unemployment_df2 = unemployment_df2.filter(like='Persons', axis=1)
unemployment_df2.drop(unemployment_df2.columns[[3,4]] ,axis=1, inplace=True)
# unemployment_df2 = unemployment_df2.drop(unemployment_df2.iloc[:, 4:27], inplace=True, axis=1)
unemployment_df2
```

```python
# Creating Year column and setting to index
date_data = time_range_df['datetime_col']
date_data_df = date_data.to_frame()

date_data_df['Year'] = pd.to_datetime(date_data_df['datetime_col']).dt.year
# date_data_df


#  merge both df
unemployment_df = pd.concat([date_data_df, unemployment_df1, unemployment_df2], axis=1)

# set index
unemployment_df = unemployment_df.set_index('Year')

#rename columns
unemployment_df.columns.values[1] = 'Australia'
unemployment_df.columns.values[2] = 'NSW'
unemployment_df.columns.values[3] = 'VIC'
unemployment_df.columns.values[4] = 'QLD'
unemployment_df.columns.values[5] = 'SA'
unemployment_df.columns.values[6] = 'WA'
unemployment_df.columns.values[7] = 'TAS'
unemployment_df.columns.values[8] = 'NT'
unemployment_df.columns.values[9] = 'ACT'

# display
unemployment_df
```

```python
unemployment_df.to_csv('./unemployment_cleaned_df.csv', index=False)
```
