```python
#Rate per 100,000 persons aged 10 years and over for the state/territory of interest.

plt.figure(1)


years = [2017,2018,2019,2020,2021]


ACT= [779.6,755.0,676.1,648.1,595.7]
NSW= [1507.6,1486.2,1454.4,1400.3,1332.6]
NT= [5702.6,5962.7,5195.9,4541.3,4343.9]
QLD= [2125.8,2019.8,1961.7,1914.3,1800.7]
SA= [2860.5,2854.2,2670.4,2260.4,1981.3]
TAS= [2279.3,2189.8,2078.5,1825.7,1745.9]
VIC= [1354.6,1349.3,1299.2,1249.6,1332.1]
WA= [1873.3,1844.5,1742.4,1636.9,1525.4]

ACT = plt.plot(years, ACT, marker="s", color="Red", linewidth=1, label="ACT")
NSW = plt.plot(years, NSW, marker="s", color="Blue", linewidth=1, label="NSW")
NT = plt.plot(years, NT, marker="s", color="Green", linewidth=1, label="NT")
SA = plt.plot(years, SA, marker="s", color="Magenta", linewidth=1, label="SA")
QLD = plt.plot(years, QLD, marker="s", color="Cyan", linewidth=1, label="QLD")
TAS = plt.plot(years, TAS, marker="s", color="Purple", linewidth=1, label="TAS")
VIC = plt.plot(years, VIC, marker="s", color="Black", linewidth=1, label="VIC")
WA = plt.plot(years, WA, marker="s", color="Yellow", linewidth=1, label="WA")


plt.title("Offender Rate per 100,000 Persons For Each State")
plt.xlabel("Years")
plt.ylabel("Number of Offences")
plt.xticks(years)
plt.grid()
plt.legend(loc='upper left')

plt.savefig('images/Graph1.png', bbox_inches='tight')

plt.show()


```
