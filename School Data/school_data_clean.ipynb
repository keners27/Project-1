```python
# Dependencies and Setup
import pandas as pd
```

```python
# Load csv with school data
school_data= pd.read_csv('Resources/school-profile-2008-2021.csv')
school_data.head()
```

```python
# Review columns for relevance
list(school_data.columns)
```

```python
# Remove superflus columns
school_data = school_data[['Calendar Year',
 'School Name',
 'State',
 'School Type',
 'ICSEA',
 'ICSEA Percentile',
 'Full Time Equivalent Teaching Staff',
 'Full Time Equivalent Enrolments']]
school_data.head()
```

```python
# Remove rows with null data
school_data.dropna(how='any',inplace=True)
```

```python
# Remove school profiles from the previous years and up to 2020 to match everyone's data
current_school_data = school_data.loc[(school_data["Calendar Year"] > 2014) & (school_data["Calendar Year"] < 2021)]
current_school_data
```

```python
# Remove school profiles with the type special
current_school_data = current_school_data.loc[school_data["School Type"] != "Special"]
current_school_data
```

```python
# Confirm output file and export data into a csv
output_csv='Output/clean_school_data.csv'

current_school_data.to_csv(output_csv,header=True,index=False)
```

#Get the ICSEA trend data

```python
# Get the yearly average ICSEA for each state
ACT_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "ACT"]
ACT_ICSEA = ACT_ICSEA.groupby(["Calendar Year"]).mean()
ACT_ICSEA = ACT_ICSEA.rename(columns={"ICSEA":"ACT"})
WA_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "WA"]
WA_ICSEA = WA_ICSEA.groupby(["Calendar Year"]).mean()
WA_ICSEA = WA_ICSEA.rename(columns={"ICSEA":"WA"})
NSW_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "NSW"]
NSW_ICSEA = NSW_ICSEA.groupby(["Calendar Year"]).mean()
NSW_ICSEA = NSW_ICSEA.rename(columns={"ICSEA":"NSW"})
VIC_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "VIC"]
VIC_ICSEA = VIC_ICSEA.groupby(["Calendar Year"]).mean()
VIC_ICSEA = VIC_ICSEA.rename(columns={"ICSEA":"VIC"})
QLD_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "QLD"]
QLD_ICSEA = QLD_ICSEA.groupby(["Calendar Year"]).mean()
QLD_ICSEA = QLD_ICSEA.rename(columns={"ICSEA":"QLD"})
NT_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "NT"]
NT_ICSEA = NT_ICSEA.groupby(["Calendar Year"]).mean()
NT_ICSEA = NT_ICSEA.rename(columns={"ICSEA":"NT"})
SA_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "SA"]
SA_ICSEA = SA_ICSEA.groupby(["Calendar Year"]).mean()
SA_ICSEA = SA_ICSEA.rename(columns={"ICSEA":"SA"})
TAS_ICSEA = current_school_data[["Calendar Year", "State","ICSEA"]].loc[current_school_data["State"] == "TAS"]
TAS_ICSEA = TAS_ICSEA.groupby(["Calendar Year"]).mean()
TAS_ICSEA = TAS_ICSEA.rename(columns={"ICSEA":"TAS"})

# Merge the Series together
summary_ICSEA = pd.concat([ACT_ICSEA,WA_ICSEA,NSW_ICSEA,VIC_ICSEA,QLD_ICSEA,NT_ICSEA,SA_ICSEA,TAS_ICSEA],axis=1)
summary_ICSEA
```

```python
output_csv='Output/summary_ICSEA_data.csv'

summary_ICSEA.to_csv(output_csv,header=True,index=True)
```
